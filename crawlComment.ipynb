```python
def writeToJson(ArrOutputData):
    y=json.dumps({"data":ArrOutputData},ensure_ascii=False).encode('utf8')
    a=json.loads(y)
    with open('comments2.json', 'w',encoding='utf8') as f:
        json.dump(a, f,ensure_ascii=False)
```
