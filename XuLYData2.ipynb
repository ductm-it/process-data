```python
import sys
import os
import path
import time
import re
import numpy as np
import json
import gzip
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
import unidecode
import joblib
```

```python

total=0
allComment=[]

# f = open("comments.json")
# data = json.load(f)
# for i in (data['data']):
#     for j in i:
#         allComment.append(j)
#         total=total+1

# f = open("comments1.json")
# data1 = json.load(f)
# for i in (data1['data']):
#     for j in i:
#         allComment.append(j)
#         total=total+1

f = open("data.json", encoding="utf8")
data2 = json.load(f)
for i in (data2['data']):
    for j in i:
        allComment.append(j)
        total=total+1
print(total)
print(allComment[0])
```

```python
contents=[]
stars=[]
for i in range(0,len(allComment)):
    contents.append(unidecode.unidecode(allComment[i]['content']).lower())
    stars.append(allComment[i]['star'])
```

```python
vectorizer = CountVectorizer()
vectorizer.fit(contents)
# print(vectorizer.vocabulary_)
X_train = vectorizer.transform(contents)
print(X_train.shape)
print(type(X_train))
print(X_train.toarray()[0])
```

```python
y_train=[]

numNega=0
numPosi=0
for i in stars:
    if(i<=3):
        y_train.append(0)
        numNega=numNega+1
    else:
        y_train.append(1)
        numPosi=numPosi+1
print(numNega)
print(numPosi)
print(len(y_train))
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
print(classifier.predict(vectorizer.transform(["hang bi loi kem chat luong"])))
```

```python
# tach ra 500 xau
y_train=[]
x_train=[]
x_train_bad=[]
x_train_good=[]

total=0
for i in range(0,len(stars)):
    if(stars[i]<=3):
        if(contents[i] not in x_train_bad):
            y_train.append(0)
            x_train_bad.append(contents[i])
            x_train.append(contents[i])
            total=total+1
    if(total==500):
        break
print(len(y_train))
print(len(list(set(x_train_bad))))

total=0
for i in range(0,len(stars)):
    if(stars[i]>3):
        if(contents[i] not in x_train_good):
            y_train.append(1)
            x_train_good.append(contents[i])
            x_train.append(contents[i])
            total=total+1
    if(total==len(x_train_bad)):
        break
print(len(y_train))
print(len(x_train))
# np.save("CleanedData/x_train.npy",x_train)
# np.save("CleanedData/y_train.npy",y_train)

print(len(list(set(x_train_good))))
```
